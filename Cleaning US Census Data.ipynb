   Using `glob`, loop through the census files available and load them into DataFrames. Then, I concatenate all of those DataFrames together into one DataFrame, called something like `us_census`.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

import glob
files = glob.glob("states*.csv")

states_list = [] #create an empty list to fill
for filename in files:
    data = pd.read_csv(filename)
    states_list.append(data)

us_census = pd.concat(states_list) #our dataframe
````

## Regex to the Rescue

5. Using regex to convert the `Income` column into a format that is ready for conversion into a numerical type.

````python
for index in range(0,len(us_census["Income"])): #using a For Loop to itterate over the elements within the column
    string = str(us_census['Income'].iat[index])
    replace_dol = string.replace('$', '') #removing $ and replacing with nothing
    replace_com = replace_dol.replace(',', '') #removing the , and also replacing with nothing
    us_census['Income'].iat[index] = replace_com

us_census["Income"] = pd.to_numeric(us_census['Income'])
us_census["Income"]
````

6. Looking at the `GenderPop` column, we would need to separate this into two separate columns, the `Men` column, and the `Women` column.

   To split the column into those two new columns, we use `str.split` and separating out those results.

````python
us_census['GenderPop'].head()

Men = [] #create an empty list to fill
Women = [] #create the other empty list to fill
for index in range(0,len(us_census["GenderPop"])): #using range, len would loop over all elements
    string = str(us_census['GenderPop'].iat[index]) #Similar to iloc, "iat" accessing a single value for a row/column pair by integer position
    replace = string.split('_')
    Men.append(replace[0])
    Women.append(replace[1])

us_census['Men'] = Men
us_census['Women'] = Women

us_census.head()
````

7. We need to convert both of the columns into numerical datatypes now.

   However, there is still an `M` or an `F` character in each entry! We should remove those before we convert into numerical datatypes.

````python
for index in range(0,len(us_census["Men"])):
    string = str(us_census['Men'].iat[index])
    replace = string.replace('M', '')
    us_census['Men'].iat[index] = replace
    
for index in range(0,len(us_census["Women"])):
    string = str(us_census['Women'].iat[index])
    replace = string.replace('F', '')
    us_census['Women'].iat[index] = replace
    
us_census['Men'] = pd.to_numeric(us_census['Men'])
us_census['Women'] = pd.to_numeric(us_census['Women'])

us_census.head()
````

8. Now we have the columns needed to make the graph and make sure the team can now visualize.

   I will use matplotlib to make a scatterplot!
   
   ```py
   plt.scatter(the_women_column, the_income_column)
   ```

````python
plt.scatter(us_census['Women'], us_census['Income'])
plt.title("Scatter Plot of Income vs. Number of Women per State")
plt.xlabel("Population of Women per State")
plt.ylabel("Income (in US Dollars)")
plt.show()
plt.clf()
````

9. It's also key double check the work as these csv files probably have `nan` values in them! Print out your column with the number of women per state to see.

   We can fill in those `nan`s by using pandas' `.fillna()` function.
   
   You have the `TotalPop` per state, and you have the `Men` per state. As an estimate for the `nan` values in the `Women` column, you could use the `TotalPop` of that state minus the `Men` for that state.
   
   Print out the `Women` column after filling the `nan` values to see if it worked!

````python
print(us_census['Women'])
us_census['Women'] = us_census['Women'].fillna(us_census['TotalPop'] - us_census['Men'])
print(us_census['Women'])
````

10. We also need to check for duplicates! Using `.duplicated()` on the `census` DataFrame to check if we have duplicate rows.

````python
us_census.duplicated(subset = us_census.columns[1:])
````

11. Dropping these duplicates using the `.drop_duplicates()` function.

````python
census = us_census.drop_duplicates(subset = us_census.columns[1:])
census
````

12. Create the scatterplot again. Now, it should be good to go!

````python
plt.scatter(census['Women'], census['Income'], color=["purple"], alpha=0.5)
plt.title("Scatter Plot of Income vs. Number of Women per State")
plt.xlabel("Population of Women per State")
plt.ylabel("Income (in US Dollars)")
plt.show()
plt.clf()
````

## Histogram of Races

13. It was also request to create histograms out of the race data we have. Bringing up the `.columns` again to see what the race categories are.

````python
census.columns
````

14. We will create a histogram for each!

    We need to firstly, convert the columns into the numerical format, and those percentage signs will have be removed.
    
    Fill the `nan` values with something that makes sense! We dropped the duplicate rows when making scatter plot, but it couldn't hurt to check for duplicates again.

````python
for race in ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']:
    for index in range(0,len(us_census)):    
        string = str(us_census[race].iat[index])
        replace = string.replace('%', '')
        if (replace == "nan"):
            replace = ""
        us_census[race].iat[index] = replace
    us_census[race] = pd.to_numeric(us_census[race])
    
us_census['Pacific'] = us_census['Pacific'].fillna(100 - us_census['Hispanic'] - us_census['White'] - us_census['Black'] - us_census['Native'] - us_census['Asian'])

census = us_census.drop_duplicates(subset = us_census.columns[1:])
census

for race in ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']:
    plt.hist(census[race],color="green", alpha=0.5)
    plt.title("Histogram of the Percentage of {} People per State".format(race))
    plt.xlabel("Percentage")
    plt.ylabel("Frequency")
    plt.grid()
    plt.show()
    plt.clf()
````
